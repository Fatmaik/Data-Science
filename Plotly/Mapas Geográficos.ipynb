```python
import plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
# Conectando com o notebook para a geração dos plots
init_notebook_mode(connected=True)
```

```python
import pandas as pd
```

```python
data = dict(type='choropleth',
           locations=['AZ', 'CA', 'NY'],
           locationmode='USA-states',
           colorscale='Portland',
           text=['Arizona', 'California', 'New Yourk'],
           z = [1.0, 2.0, 3.0],
           colorbar={'title': 'Titulo da barra de cores'})
```

```python
data
```

```python
layout=dict(geo={'scope':'usa'})
```

```python
choromap = go.Figure(data=[data], layout=layout)
```

```python
iplot(choromap)
```

```python
# Exemplo 2
df = pd.read_csv('2011_US_AGRI_Exports')
```

```python
df.head(10)
```

```python
data = dict(type='choropleth',
    colorscale='YlOrRd',
    locations=df['code'],
    z=df['total exports'],
    marker=dict(line=dict(color='rgb(0, 0, 0)', width=3)),
    locationmode='USA-states',
    text=df['state'],
    colorbar={'title': 'Titulo da barra de cores'})
```

```python
layout = dict(title='Exportações agricolas de 2011', geo=dict(scope='usa', showlakes=True, lakecolor='rgb(85, 173, 240)'))
```

```python
layout
```

```python
choromap = go.Figure(data=[data], layout=layout)
iplot(choromap)
```

```python
# Exemplo 3
df = pd.read_csv('2014_world_GDP')
df.head()
```

```python
data = dict(type='choropleth',
           locations=df['CODE'],
           z=df['GDP (BILLIONS)'],
           text=df['COUNTRY'],
           colorbar={'title': 'PIB em bilhões de dólores'})
```

```python
layout = dict(title='PIB mundo 2014', geo=dict(showframe=False, projection={'type': 'mercator'}))
```

```python
choromap = go.Figure(data=[data], layout=layout)
```

```python
iplot(choromap)
```
