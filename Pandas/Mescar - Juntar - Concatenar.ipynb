## Concatenação

Basicamente é a unicão de dois ou mais dataFrames, mas se atente que para<br>
ocorrer uma concatenação bem executada, a dimensão dos DataFrames deve ser igual ex:3x3

```python
pd.concat([df1, df2, df3])
```

```python
pd.concat([df1, df2, df3], axis=1)
```

## Mesclar

Fazer a união de duas dataFrames ultilizando uma coluna em comum

```python
esquerda = pd.DataFrame({
    'A':['A0', 'A1', 'A2', 'A3'],
    'B':['B0', 'B1', 'B2', 'B3'],
    'Key':['Key0', 'Key1', 'Key2', 'Key3']    
})
```

```python
direita = pd.DataFrame({
    'C':['C0', 'C1', 'C2', 'C3'],
    'D':['D0', 'D1', 'D2', 'D3'],
    'Key':['Key0', 'Key1', 'Key2', 'Key3']    
})
```

```python
esquerda
```

```python
direita
```

```python
pd.merge(esquerda, direita, how='inner', on='Key')
```

## Juntar

União de dois DataFrames que possivelmente possuem valores correspondende por indice

```python
direita = pd.DataFrame({
    'A':['A0', 'A1', 'A2'],
    'B':['B0', 'B1', 'B2']  
}, index=[0,1,2])
```

```python
esquerda = pd.DataFrame({
    'C':['C0', 'C2', 'C3'],
    'D':['D0', 'D2', 'D3']  
}, index=[0,1,2])
```

```python
direita
```

```python
esquerda
```

```python
direita.join(esquerda)
```

```python
esquerda.join(direita)
```

```python
esquerda.join(direita, how='outer')
```
